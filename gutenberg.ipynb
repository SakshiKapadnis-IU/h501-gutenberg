## Section 2: Load and Explore Data with Pandas

Let's create a sample dataset and explore its structure using pandas methods.

````python
# Create sample data using our custom module
df = create_sample_data(n_rows=200)

# Display the first few rows
print("First 5 rows of the dataset:")
print(df.head())
print()

# Display basic information about the dataset
print("Dataset Information:")
df.info()
print()

# Display summary statistics
print("Summary Statistics:")
print(df.describe())
````

## Section 3: Data Cleaning and Preprocessing

We'll demonstrate various data cleaning operations using pandas.

````python
# Add some missing values to demonstrate cleaning
df_with_missing = df.copy()
df_with_missing.loc[5:10, 'value'] = np.nan
df_with_missing.loc[15:20, 'count'] = np.nan

print("Dataset shape before cleaning:", df_with_missing.shape)
print("Missing values before cleaning:")
print(df_with_missing.isnull().sum())
print()

# Clean the data using our module
df_clean = clean_data(df_with_missing)

print("Dataset shape after cleaning:", df_clean.shape)
print("Missing values after cleaning:")
print(df_clean.isnull().sum())
print()

# Display the cleaned data summary
summary = summarize_data(df_clean)
print("Data Summary:")
print(f"Rows: {summary['rows']}")
print(f"Columns: {summary['columns']}")
print(f"Missing values: {summary['missing_values']}")
````

````python
# Demonstrate filtering data
print("Example 1: Filter data for category 'A'")
category_a = filter_data(df, 'category', 'A')
print(f"Rows with category 'A': {len(category_a)}")
print(category_a.head())
print()

# Demonstrate aggregation
print("Example 2: Aggregate data - Average value by category")
avg_by_category = aggregate_data(df, 'category', 'value', 'mean')
print(avg_by_category)
print()

print("Example 3: Aggregate data - Sum of count by category")
sum_by_category = aggregate_data(df, 'category', 'count', 'sum')
print(sum_by_category)
````

## Section 6: PEP-8 Style Guidelines

All code in this project follows PEP-8 style guidelines. Here are the key conventions applied:

### 1. Naming Conventions
- **Variables and Functions**: `snake_case` (e.g., `create_sample_data`, `df_clean`)
- **Classes**: `PascalCase` (e.g., `DataFrame`)
- **Constants**: `UPPER_SNAKE_CASE`
- **Private functions**: Leading underscore (e.g., `_helper_function`)

### 2. Code Organization
- **Imports**: Organized at the top (standard library, third-party, local)
- **Line length**: Maximum 79 characters for code, 72 for comments
- **Blank lines**: Two blank lines between module-level functions/classes
- **Indentation**: 4 spaces (no tabs)

### 3. Documentation
- **Module docstrings**: Describe the module's purpose
- **Function docstrings**: Follow NumPy style with Parameters, Returns, Examples
- **Inline comments**: Explain complex logic, use sparingly

### 4. Spacing
- **Around operators**: `x = y + 1` (spaces around `=` and `+`)
- **After commas**: `function(arg1, arg2, arg3)`
- **No space before colons**: `if condition:` not `if condition :`
